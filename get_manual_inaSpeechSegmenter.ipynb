```python
from scipy import stats
import pandas as pd
pd.set_option('display.max_columns', None)

df = pd.read_csv("./csv/df-all-features.csv")
df = df.sample(n=10, random_state=4)
display(df)
```

```python
manual_annotations = {
    1: [('female',0.0,30.0)],
    2: [('noise',0.0,9.0),('male',9.0,30.0)],
    3: [('female',0.0,30.0)],
    4: [('female',0.0,30.0)],
    5: [('male',0.0,30.0)],
    6: [('music',0.0,17.0),('noEnergy',17.0,18.0),('female',18.0,30.0)],
    7: [('male',0.0,15.0),('noEnergy',15.0,16.0),('male',16.0,30.0)],
    8: [('music',0.0,1.0),('male',1.0,30.0)],
    9: [('music',0.0,10.0),('male',10.0,30.0)],
    10: [('noEnergy',0.0,2.0),('noise',2.0,10.0),('male',10.0,18.0),('noise',18.0,23.0),('male',23.0,30.0)]
}


counter = 1
for index, row in df.iterrows():

    tuple_list = manual_annotations[counter]
    
    for attribute in ["female-manual", "male-manual", "music-manual", "noEnergy-manual", "noise-manual"]:
        sum_seconds = 0.0
        for item in tuple_list:
            target = attribute.replace("-manual","")
            if target == item[0]:
                sum_seconds += item[2] - item[1]
        df.loc[index, attribute] = sum_seconds

    counter += 1
    
display(df)
print(stats.pearsonr(x=df["female"], y=df["female-manual"]))
print(stats.pearsonr(x=df["male"], y=df["male-manual"]))
```
